```python
import numpy as np
import math
from sklearn.metrics import mean_squared_error, mean_absolute_error, log_loss
import matplotlib.pyplot as plt
np.random.seed(9)
```

# MSE

```python
def my_mean_squared_error(observe: list, y_pred: list):
    """Моя реализация MSE руками"""
    if not isinstance(observe, list) or not isinstance(y_pred, list):
        raise TypeError("Observances and target must be list type.")
    if len(observe) != len(y_pred):
        raise ValueError(f"Target have {len(observe)} item, but y_pred have {len(y_pred)}.")
    
    c_item = 1/len(observe)
    minus = []
    for i in range(len(observe)):
        minus.append((observe[i] - y_pred[i])**2)
    return c_item * sum(minus)
```

```python
print(f"SkLearn MSE: {mean_squared_error([3, 2, 1], [1, 1, 1])}")
print(f">>>> My MSE: {my_mean_squared_error([3, 2, 1], [1, 1, 1])}")
```

# MAE

```python
def my_mean_absolute_error(observe: list, y_pred: list):
    """Моя реализация MAE руками"""
    if not isinstance(observe, list) or not isinstance(y_pred, list):
        raise TypeError("Observances and target must be list type.")
    if len(observe) != len(y_pred):
        raise ValueError(f"Target have {len(observe)} item, but y_pred have {len(y_pred)}.")
    
    c_item = 1/len(observe)
    minus = []
    for i in range(len(observe)):
        minus.append(abs((observe[i] - y_pred[i])))
    return c_item * sum(minus)
```

```python
print(f"SkLearn MAE: {mean_absolute_error([3.11, 22, 7], [4, 21.3, 5])}")
print(f">>>> My MAE: {my_mean_absolute_error([3.11, 22, 7], [4, 21.3, 5])}")
```

# Binary cross-entropy

```python
def my_binary_cross_entropy(input: list, target: list,):
    """Моя реализация бинарной кросс-энтропии" руками"""
    if not isinstance(input, list) or not isinstance(target, list):
        raise TypeError("Observances and target must be list type.")
    if len(input) != len(target):
        raise ValueError(f"Target have {len(input)} item, but y_pred have {len(target)}.")
    
    bce_sum = sum([target[i] * math.log(input[i]) + (1-target[i]) * math.log(1-input[i]) for i in range(len(input))])
    
    return -bce_sum / len(input)
```

```python
input_values = [0.99999999, 0.000001, 0.89999999]
target_values = [1, 0, 1]

print(f"SkLearn Бинарная кросс-энтропия: {log_loss(target_values, input_values)}")
print(f">>> Моя Бинарная кросс-энтропия: {my_binary_cross_entropy(input=input_values, target=target_values)}")
```

# GD

```python
f = lambda x: x**3 -2 * x**2 + 2
x = np.linspace(-1, 2.5, 1000)
plt.plot(x, f(x))
plt.grid()
```

```python
f_prime = lambda x: 3 * x**2 - 4 * x 
x = np.linspace(-1, 2.5, 1000)
plt.plot(x, f_prime(x))

plt.ylim([-2, 3])
plt.grid()
```

```python
lr = 0.1
pos = [2]
n_epoch = 15
```

```python
for i in range(n_epoch):
    current_p = pos[-1]
    grad = f_prime(current_p)
    new_point = current_p - lr * grad
    pos.append(new_point)
```

```python
y_points = [f(x) for x in pos]
```

```python
plt.plot(x, f(x))
plt.plot(pos, y_points, "ro")
```
